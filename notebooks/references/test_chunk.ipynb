```python
from lionagi.tools.types import chunk

url = "https://arxiv.org/pdf/2501.18438"
title = "O3-MINI VS DEEPSEEK-R1: WHICH ONE IS SAFER?"
date = "2025-01-31"
author = "Aitor Arrieta"

chunks = chunk(
    url,
    reader_tool="docling",
    metadata={
        "url": url,
        "title": title,
        "date": date,
        "author": author,
    },
    chunk_size=2000,
    overlap=0.05,
)
```

```python
len(chunks)
```

```python
print(chunks[0].content)
```

```python
chunks = chunk(
    url,
    reader_tool="docling",
    chunk_by="tokens",
    metadata={
        "url": url,
        "title": title,
        "date": date,
        "author": author,
    },
    chunk_size=300,
    overlap=0.05,
)
```

```python
len(chunks)
```

```python
print(chunks[0].content)
```
