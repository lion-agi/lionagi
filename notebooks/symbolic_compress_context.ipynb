```python
from pathlib import Path

fp = Path.cwd() / "data" / "rust" / "example_rust.rs"
output_path = Path.cwd() / "data" / "rust" / "example_rust_compressed.txt"
```

```python
from lionagi import iModel
from lionagi.libs.token_transform.types import (
    symbolic_compress_context,
    TokenMappingTemplate,
)

chat_model = iModel(model="openrouter/google/gemini-2.0-flash-001")
compress_model = iModel(model="openai/gpt-4o-mini", temperature=0.1)

results = await symbolic_compress_context(
    chat_model=chat_model,
    compress_model=compress_model,
    url_or_path=fp,
    compress=True,
)
```

```python
print(results)
```

```python
output_fp = await symbolic_compress_context(
    chat_model=chat_model,
    compress_model=compress_model,
    url_or_path=fp,
    encode_token_map=TokenMappingTemplate.RUST_CHINESE,
    encode_output=True,
    max_concurrent=5,
    throttle_period=1,
    output_path=output_path,
    verbose=False,
)

print(f"Output written to: {output_fp}")

text = output_fp.read_text()
print(text)
```
