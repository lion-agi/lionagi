```python
import lionagi as li

instruction = "Similarity between pizza and quantum mechanics, in few words, use rhymes"
```

### Ollama

```python
imodel = li.iModel(
    provider="ollama",
    model="llama3",
)
```

```python
branch = li.Branch(imodel=imodel)
a = await branch.chat(instruction)
print(a)
```

### mlx_lm

```python
imodel = li.iModel(model="mlx-community/Qwen1.5-1.8B-Chat-4bit", provider="mlx")
```

```python
branch = li.Branch(imodel=imodel)
a = await branch.chat(instruction, max_tokens=100)
print(a)
```

### Transformers

```python
imodel = li.iModel(
    provider="transformers",
    model="Columbia-NLP/gemma-2b-zephyr-dpo",
)
```

```python
branch = li.Branch(imodel=imodel)
a = await branch.chat(instruction, max_tokens=100)
print(a)
```
